```python
subjects = ["Toán","Ngữ văn","Địa lý","Tiếng anh","Vật lí","Hóa học","Lịch sử","Sinh học"]
for i in range(len(df["DIEM_THI"])):
    for j in range(len(subjects)):
        if subjects[j] in df.loc[i,'DIEM_THI']:
            begin = df.loc[i,'DIEM_THI'].find(subjects[j])
            end = begin+len(subjects[j])+2
            print(df.loc[i,"DIEM_THI"][end:end+6])
            df.loc[i,subjects[j]] = df.loc[i,"DIEM_THI"][end:end+6]
            
```
